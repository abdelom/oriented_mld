```python
def replicate(nb=30, oriented=True, mrf=None, block=False):
    params = {"sample_size": 10, "Ne": 1, "ro": 8e-5, "mu": 8e-4, "Tau": 1.0, "Kappa": 10.0 , "length": int(1e6)}
    length_distribution = np.array([0.0] * 26)
    for i in range(10):
        _, breakpoints, variants = om.msprime_simulate_variants(params)
        if mrf is None:
            mld = om.detect_internal_incompatibilities(variants, oriented=oriented, thresold=20)
            if block:
                mld = om.incompatible_blocks_2(variants, oriented, mld)
                mld = om.incompatible_blocks_2(variants, oriented, mld)
            mld = om.choping(mld, variants, int(1e6) )
        else:
            mld = breakpoints
        length_distribution += om.block_length(mld, variants)
    return length_distribution / nb

data = pd.DataFrame([*zip(replicate(oriented=False),
                    replicate(),
                    replicate(mrf="m"),
                    replicate(oriented=False, block=True), 
                    replicate(block=True))], columns=["non_oriented", "oriented","mrf", "non_orientde_2", "oriented_2"])
```
